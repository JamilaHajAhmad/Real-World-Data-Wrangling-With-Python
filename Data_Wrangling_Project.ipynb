### Quality Issue 1:

```python
# Inspecting the mental_health_df visually using head()
mental_health_df.head()

# Inspecting the mental_health_df visually using sample()
mental_health_df.sample(5)
```

```python
# Inspecting the mental_health_df programmatically
mental_health_df.info()
```

Issue and justification: 
The issue with the selfEmployed column comes from how missing values are represented in the dataset.<br> The info() method only reports missing values as NaN (Not a Number) or None in pandas, but sometimes missing data is represented in alternative ways,such as:

**Placeholder Values**: Missing values could be represented by strings like 'NA', 'N/A', 'Unknown', 'None', or other similar placeholders<br> that pandas doesn't recognize as null automatically.

**Whitespace**: Missing entries could be represented by empty strings ('' or ' '), which info() does not interpret as missing.<br>
So I've to solve this and get back to the main issue.

```python
# Using unique() to know the representation of null values 
mental_health_df['selfEmployed'].unique()

# Replacing null values with NAN
import numpy as np
missing_values = [''] 
mental_health_df['selfEmployed'] = mental_health_df['selfEmployed'].replace(missing_values, np.nan)

# Rechecking the unique values of selfEmployed
mental_health_df['selfEmployed'].unique()

# Re-inspecting the mental_health_df programmatically
mental_health_df.head()
```

Now, after the missing values representation issue was solved, we can now justify the main issue.

```python
# Inspecting the mental_health_df programmatically
mental_health_df.info()
```

The *quality issue* here is **completeness**, as we can see there are missing values in *selfEmployed* column.

## 3. Clean data
In this section, we discuss the steps taken to clean and prepare the data for analysis.<br> Cleaning the data is an essential part of any data science project, as raw data often contains errors, inconsistencies, and irrelevant information.<br> The goal of data cleaning is to ensure that the data is tidy, accurate, and formatted correctly for analysis.

```python
# Make copies of the datasets to ensure the raw dataFrames are not impacted
clean_performance_df = performance_df.copy()
clean_mental_health_df = mental_health_df.copy()
clean_new_performance_df = new_performance_df.copy()
```

### **Quality Issue 1:**

```python
# Apply the cleaning strategy
clean_mental_health_df['selfEmployed'].fillna('Unknown', inplace=True)
```

```python
# Validate the cleaning was successful
clean_mental_health_df.head()
```

Justification:

I feel that the missing values are significant and should not be replaced with the most frequent category.<br>
This approach acknowledges that the missing data might carry some meaning, and it allows to retain the<br> distinction between missing and known values, which could be useful for later analysis.

### **Quality Issue 2:**

```python
# Apply the cleaning strategy
teacher_quality_mode = clean_performance_df['Teacher_Quality'].mode()[0]
parental_education_mode = clean_performance_df['Parental_Education_Level'].mode()[0]
distance_from_home_mode = clean_performance_df['Distance_from_Home'].mode()[0]
clean_performance_df['Teacher_Quality'].fillna(teacher_quality_mode, inplace=True)
clean_performance_df['Parental_Education_Level'].fillna(parental_education_mode, inplace=True)
clean_performance_df['Distance_from_Home'].fillna(distance_from_home_mode, inplace=True)
```

```python
# Validate the cleaning was successful
print(sum(clean_performance_df['Teacher_Quality'].isnull()))
print(sum(clean_performance_df['Parental_Education_Level'].isnull()))
print(sum(clean_performance_df['Distance_from_Home'].isnull()))
clean_performance_df.info()
```

Justification: 
I chose to fill the missing values with the mode of the respective column, as this approach is generally considered a good strategy for dealing with categorical variables.<br>
This approach maintains the integrity of the data and allows for meaningful analysis.

### **Tidiness Issue 1:**

```python
# Apply the cleaning strategy
person_df = mental_health_df[['id','timestamp','gender','country','occupation','selfEmployed']]
clean_mental_health_df = mental_health_df[['id','occupation','familyHistory','treatment','daysIndoors','growingStress','changesHabits','mentalHealthHistory','moodSwings','copingStruggles','workInterest','socialWeakness','mentalHealthInterview','careOptions']]
```

```python
# Setting the id to be the index
person_df.set_index('id', inplace=True)
clean_mental_health_df.set_index('id', inplace=True)
```

```python
# Get a clean_mental_heath_df for students only
clean_mental_health_df = clean_mental_health_df.query('occupation == "Student"')

# Drop the occupation column as it is not needed anymore (since it is already present in the person_df)
clean_mental_health_df.drop('occupation', inplace=True, axis=1)
```

```python
# Validate the cleaning was successful
clean_mental_health_df.sample(10)
```

Justification: 

By splitting the tables, we maintain clear boundaries between person-related data and mentalHealth-related data, adhering to the tidy data principle: each type of observational unit should form a table.

## 5. Answer the research question

### **5.1:** Define and answer the research question 


*Research question:* 
    How do numerical performance metrics (attendance, hours studied, exam score) vary among male and female students, and how are these metrics correlated?

```python
# Visual 1

import matplotlib.pyplot as plt

clean_student_performance_df.hist(column=['Attendance', 'Hours_Studied', 'Exam_Score'], 
                                by='Gender', 
                                figsize=(10, 8), 
                                bins=10, 
                                layout=(3, 2), 
                                sharex=False, 
                                sharey=False,
                                legend=True)
plt.tight_layout()
plt.show()

```

*Answer to research question:* 

*Attendance*: Both genders have similar attendance distributions, with most students in the middle range.

*Hours Studied*: Females slightly dominate lower study hours, while males appear more evenly distributed.

*Exam Scores*: Males and females show comparable peaks around the mid-range, with slightly higher male variability.


```python
# Visual 2

# Plot the pie chart for each column by 'Gender'
clean_student_performance_df.groupby('Gender')[['Attendance', 'Hours_Studied', 'Exam_Score']]\
    .sum()\
    .plot.pie(subplots=True, figsize=(11, 9), autopct='%1.1f%%', legend=True, startangle=90, labels=['F','M'])
plt.show()
```

*Answer to research question:* 

For Attendance, Hours_Studied, and Exam_Score, the distribution of males and females appears to be almost equal, with males representing about 57.8% and females 42.2% for each category.

There is a consistent trend across all three charts where males (M) have a slightly higher percentage than females (F) in all categories: Attendance, Hours Studied, and Exam Score.

### Conclusion 

Based on the current data shown by these pie charts and histograms, gender does not seem to have a significant effect on the student's performance (in terms of attendance, hours studied, and exam score) when considering their mental health outcomes.

However, more in-depth analysis, such as looking at correlations or further statistical testing, would be required for a deeper understanding.
